```python
def create_graph(grapname:str):    
    unique_topics = gh_df['topic'].unique().tolist()
    mean_element = []
    for unique_topic in unique_topics:
        element = gh_df.loc[gh_df['topic'] == unique_topic, grapname].mean()
        mean_element.append((element))
    mean_element
    import matplotlib.pyplot as plt
    plt.figure(figsize=(50, 15))
    plt.bar(unique_topics, mean_element)
    plt.xlabel('Topic',fontsize =20)
    plt.ylabel('Mean '+ grapname,fontsize =20)
    plt.xticks(rotation=45, ha='right', fontsize=30)
    plt.yticks(rotation=45, ha='right', fontsize=30)
    plt.show()
```

```python
import folium
from geopy.geocoders import Nominatim
def location_on_map(cities,map_name:str):    
    
    
    
    # Erstelle eine Karte
    map_cities = folium.Map(location=[0, 0], zoom_start=2)
    
    # F√ºge Marker f√ºr jeden Stadtnamen hinzu
    geolocator = Nominatim(user_agent="my_geocoder")
    
    for city in cities:
        location = geolocator.geocode(city)
        if location:
            folium.Marker([location.latitude, location.longitude], popup=city).add_to(map_cities)
    
    # Speichere die Karte als HTML-Datei
    map_cities.save(map_name+'.html')
```

```python
RestAPI_DF = pd.read_csv('GITHUB_RestAPI.csv')
RestAPI_DF['created_at'] = pd.to_datetime(RestAPI_DF['created_at'])
RestAPI_DF['updated_at'] = pd.to_datetime(RestAPI_DF['updated_at'])
RestAPI_DF
```

```python
from datetime import datetime

RestAPI_DF['created_at'] = pd.to_datetime(RestAPI_DF['created_at'], format='%Y-%m-%d %H:%M:%S%z')
given_date = datetime.strptime('2017-12-11 17:51:56+00:00', '%Y-%m-%d %H:%M:%S%z')
current_datetime = datetime.now(given_date.tzinfo)
RestAPI_DF['Days_Created'] = (current_datetime - RestAPI_DF['created_at']).dt.days
RestAPI_DF['Days_Created']

```

```python
df_subset = RestAPI_DF[['followers', 'Days_Created']]
plt.scatter(df_subset['Days_Created'], df_subset['followers'])
plt.title('Follower -Experience')
plt.ylabel('Follower ')
plt.xlabel('Days_Created')
plt.show()
```

```python
RestAPI_DF['updated_at'] = pd.to_datetime(RestAPI_DF['updated_at'], format='%Y-%m-%d %H:%M:%S%z')
given_date = datetime.strptime('2017-12-11 17:51:56+00:00', '%Y-%m-%d %H:%M:%S%z')
current_datetime = datetime.now(given_date.tzinfo)
RestAPI_DF['Days_Updated'] = (current_datetime - RestAPI_DF['updated_at']).dt.days
RestAPI_DF['Days_Updated']

df_subset = RestAPI_DF[['followers', 'Days_Updated']]
plt.scatter(df_subset['Days_Updated'], df_subset['followers'])
plt.title('Follower -Hard Works')
plt.ylabel('Follower ')
plt.xlabel('Days_Updated')
plt.show()
```
