```python
import pandas as pd
gh_df = pd.read_csv('Github_data.csv')
gh_df['watch'] = gh_df['watch'].str.replace('k', '').astype(float) * 1000
gh_df['star'] = gh_df['star'].str.replace('k', '').astype(float) * 1000
gh_df['fork'] = gh_df['fork'].str.replace('k', '').astype(float) * 1000
gh_df
```

```python
top_10_forked_projects = gh_df.nlargest(10,'fork')
top_10_forked_projects
```

```python
top_10_star_projects = gh_df.nlargest(10,'star')
top_10_star_projects
```

```python
top_10_watch_projects = gh_df.nlargest(10,'watch')
top_10_watch_projects
```

```python
unique_topics = gh_df['topic'].unique().tolist()
mean_fork = []
for unique_topic in unique_topics:
    fork = gh_df.loc[gh_df['topic'] == unique_topic, 'fork'].mean()
    mean_fork.append((fork))
mean_fork
import matplotlib.pyplot as plt
plt.figure(figsize=(50, 15))
plt.bar(unique_topics, mean_fork)
plt.xlabel('Topic',fontsize =20)
plt.ylabel('Mean Fork',fontsize =20)
plt.show()
```

```python
#unique_topics = gh_df['topic'].unique().tolist()
mean_star = []
for unique_topic in unique_topics:
    fork = gh_df.loc[gh_df['topic'] == unique_topic, 'star'].mean()
    mean_star.append((fork))
mean_fork
import matplotlib.pyplot as plt
plt.figure(figsize=(50, 15))
plt.bar(unique_topics, mean_star)
plt.xlabel('star',fontsize =20)
plt.ylabel('Mean star',fontsize =20)
plt.show()
```
