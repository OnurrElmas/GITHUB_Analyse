```python
def location_on_map(cities,map_name:str):    
    import folium
    from geopy.geocoders import Nominatim
    
    
    # Erstelle eine Karte
    map_cities = folium.Map(location=[0, 0], zoom_start=2)
    
    # F√ºge Marker f√ºr jeden Stadtnamen hinzu
    geolocator = Nominatim(user_agent="my_geocoder")
    
    for city in cities:
        location = geolocator.geocode(city)
        if location:
            folium.Marker([location.latitude, location.longitude], popup=city).add_to(map_cities)
    
    # Speichere die Karte als HTML-Datei
    map_cities.save(map_name+'.html')
```
